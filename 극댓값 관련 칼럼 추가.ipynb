```python
import pandas as pd

path='/content/drive/Shareddrives/캡스톤디자인/동서기공 RUL data/max_points.json'

data=pd.read_json(path)
```

```python
data
```

```python
import statistics
data['src_median']=''
for i in range(len(data)):
    data['src_median'][i] = statistics.median(data['max_points'][i])
```

```python
data['src_median']
```

```python
data['src_Q1']=''
data['src_Q3']=''
data['src_sd']=''

for i in range(len(data)):
    data['src_Q1'][i] = pd.Series(data['max_points'][i]).quantile(0.25)
    data['src_Q3'][i] = pd.Series(data['max_points'][i]).quantile(0.75)
    data['src_sd'][i] = np.var(data['max_points'][i])
```

```python
data.to_json('/content/drive/Shareddrives/캡스톤디자인/동서기공 RUL data/max_points_plus.json')
```
